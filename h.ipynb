```python
df = df.dropna(subset=['OverAll Score'])
```

```python

# Dividir la columna en dos y asignar los resultados a nuevas columnas 'female' y 'male'
df[['female', 'male']] = df['Female:Male Ratio'].str.split(':', expand=True)

# Convertir las nuevas columnas a tipo numérico
df['female'] = pd.to_numeric(df['female'], errors='coerce')
df['male'] = pd.to_numeric(df['male'], errors='coerce')
```

```python
df = df.drop(columns=["Female:Male Ratio"])
```
